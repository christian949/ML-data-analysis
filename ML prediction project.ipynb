```python
data = pd.read_csv('csvfile.csv')
```

```python
data2 = data[['Medlemsnummer (Virksomhed) (Virksomhed)', "Type", "Key Account Manager (Virksomhed) (Virksomhed)", "Indmeldt Via", "Marked (Virksomhed) (Virksomhed)", "Strategiområde", "Indmeldelsesdato", "Udmeldelsesdato", "Udmeldelsesgrund", 'Udmeldt - BI']].copy()
```

```python
data3 = data2[['Medlemsnummer (Virksomhed) (Virksomhed)', "Type", "Marked (Virksomhed) (Virksomhed)", "Indmeldelsesdato", 'Udmeldt - BI']].copy()
```

```python
data3.head()
```

```python
#split data frame in two: one for the target variable attrition and one for the remaining features 
Y = data3['Udmeldt - BI']
X = data3.drop(['Udmeldt - BI'],axis=1)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
from xgboost import XGBClassifier
#set variable for XGBClassifier
clf = XGBClassifier()

#fit model
clf.fit(X_train,y_train)

#set variable for classifier prediction
predictions = clf.predict(X_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions) #test accuracy score 
```

```python
#calculate f-scores for the selected features 
df_score = pd.DataFrame.from_dict(clf.get_booster().get_fscore(), orient='index', columns=['F_Score'])
df_score.sort_values(by='F_Score')
```
